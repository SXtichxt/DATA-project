```python
import pandas as pd
import numpy as np
import re #regular expression
from sklearn.feature_extraction.text import TfidfVectorizer #tranform twxt to num
from sklearn.model_selection import train_test_split #auto split train&test
from sklearn.linear_model import LogisticRegression #trend sentiment
from sklearn.metrics import accuracy_score #measure accuracy score
from sklearn.metrics import confusion_matrix #matrix
print("NLP Pipeline")
```

```python
df = pd.read_csv(r"customer_feedback_cleaned.csv")
```

```python
df
```

```python
#prepare data
df[['feedback','sentiment']]
```

```python
new_data = pd.DataFrame({
    'feedback': [
        "ได้รับสินค้าแล้ว",
        "จัดส่งตามกำหนด",
        "ยังไม่ได้ลองใช้งาน"
    ],
    "sentiment":["neutral","neutral","neutral"]
})

df = pd.concat([df, new_data], ignore_index=True) 
```

```python
# Text cleansing
df['feedback_clean'] = df['feedback'].str.lower()
```

```python
df['feedback_clean'] = df['feedback_clean'].apply(
    lambda x: re.sub(r'[^ก-๙a-zA-Z0-9\s]', '',x) #tranform all text text that is not thai,eng,number,spacebar   
)
```

```python
#Vectorization TF-IDF

vectorizer = TfidfVectorizer(
    max_features = 1000,
    ngram_range = (1,2)
)

X = vectorizer.fit_transform(df['feedback_clean'])
y = df['sentiment']
```

```python
#Train/test

X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size = 0.25, 
    random_state = 42,
    stratify = y
)
```

```python
#Build model baseline
model = LogisticRegression(max_iter = 1000)
model.fit(X_train, y_train)
```

```python
model.classes_
```

```python
#Logistic Regression with L2 regularization trained on TF-IDF features using the LBFGS solver for multi-class sentiment classification.
```

```python
#predict test
y_pred = model.predict(X_test)
```

```python
#measure accuracy
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
#confusion Matrix
confusion_matrix(y_test, y_pred) 
```
